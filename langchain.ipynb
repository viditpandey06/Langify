```python
from langchain_openai import OpenAI
```

```python
import os
import API_KEY_VALUES
os.environ["OPEN_API_KEY"]=API_KEY_VALUES.OPEN_AI_KEY
```

```python
llm=OpenAI(openai_api_key=os.environ["OPEN_API_KEY"],temperature=0.6)
```

Temprature variable is used the set the creativity of the answer generated by the model. Ranges between 0 to 1.

```python
text="What is the meaning of life"
print(llm.predict(text))
```

```python
llm.predict("can you tell me the meaning of life?")
```
